```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(0,10,100)
plt.rcParams["figure.figsize"] = 10,5
fig = plt.figure()
plt.plot(x, np.sin(x), "-", c = "red")
plt.plot(x, np.cos(x), "--", c = "green")
plt.grid()
plt.show()
```

```python
plt.plot(x, np.tan(x), "-", c = "orange")
```
