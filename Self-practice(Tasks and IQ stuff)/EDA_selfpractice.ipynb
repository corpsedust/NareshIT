```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data
np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=10, size=1000)  # Symmetric
right_skewed_data = np.random.exponential(scale=10, size=1000)  # Right-skewed
left_skewed_data = -np.random.exponential(scale=10, size=1000) + 50  # Left-skewed

# Create figure
plt.figure(figsize=(12, 5))

# Plot Normal Distribution
plt.subplot(1, 3, 1)
sns.kdeplot(normal_data, fill=True, color="blue")
plt.title("Normal (Symmetric)")

# Plot Right-Skewed Distribution
plt.subplot(1, 3, 2)
sns.kdeplot(right_skewed_data, fill=True, color="green")
plt.title("Right-Skewed (Positive)")

# Plot Left-Skewed Distribution
plt.subplot(1, 3, 3)
sns.kdeplot(left_skewed_data, fill=True, color="red")
plt.title("Left-Skewed (Negative)")

# Show plots
plt.tight_layout()
plt.show()
```

```python
# Since Price and Kilometers driven data are highly skewed, we apply log
# transformation to to normalize them
# Function for log transformation of a column

def log_transform(data, col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname] + 1)
        else:
            data[colname +'_log'] = np.log(data[colname])
    data.info()
```

```python
df["P_log"] = np.log(df["Price"])
df.head()
```

```python
df["KMD_log"] = np.log(df["Kilometers_Driven"])
df.head()
```
