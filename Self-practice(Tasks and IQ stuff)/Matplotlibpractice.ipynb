```python
x1 = np.linspace(0,10,100)

fig = plt.figure()

plt.plot(x1,np.sin(x1), "-")
plt.plot(x1,np.cos(x1), "--")
```

```python
fig1 = plt.subplot(121)
fig2 = plt.subplot(122)


fig1.plot(x1,np.sin(x1), c = "red")
fig2.plot(x1,np.cos(x1),c = "cyan")
```

```python
fig3 = plt.subplot(223)
fig4 = plt.subplot(224)

fig3.plot(x1,np.tan(x1),c = "violet")
fig4.plot(x1,np.log10(x1),c = "pink")

```
