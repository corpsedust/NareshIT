### Average of Average Prices in each region over the years

- filter out entries corresponding to each region
- in a new data frame with one column as region, record :
    - max_avg_price
    - min_avg_prie
    - avg_avg_price

```python
regions = avocado.region.unique()
dfT = pd.DataFrame(columns = ["Region", "avg_avg_price", "max_avg_price", "min_avg_price"])
for i in regions:
    filterdf = avocado[avocado["region"] == i]
    values = filterdf[["AveragePrice"]].agg(["min", "max", "mean"])
    dfT = pd.concat([dfT, pd.DataFrame({"Region": [i], 
            "avg_avg_price": [values.loc["mean", "AveragePrice"]], 
            "max_avg_price": [values.loc["max", "AveragePrice"]], 
            "min_avg_price": [values.loc["min", "AveragePrice"]]})], ignore_index=True)
dfT
```

```python
#AVerageprice overthe years
plt.figure(figsize=(12,10))
data = dfT.sort_values("avg_avg_price", ascending= False).head(20)
plt.bar(x = data.Region, height=data.avg_avg_price)
plt.xticks(rotation='vertical')
plt.show()
```
